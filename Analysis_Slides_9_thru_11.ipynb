# US Computer and Mathematical Employment Analysis
----

### Analysis
* At nation level in 2017, office and administrative support, sales, and food preparation and serving are the 3 largest major occupations. Computer and mathematical occupations rank No. 14 among 22 major occupation categories. 
  Inside computer and mathematical occupations, software developers, computer user support specialists and computer system analysts occupations had the highest number of employees.
* At state level in 2017, California, Texas, and New York are the top 3 states had the most people working in computer and mathematical related occupations.
---


```python
#import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#dependencies for plotly
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True) #use plotly for offline plot and inside notebook
```

```python
#file source
excel_path = "Resources/state_M2017_dl.xlsx"

#read the excel file
state17_df = pd.read_excel(excel_path, index_col=None)
state17_df.head()
```

```python
state17_df.keys()
```

```python
#delete some columns
state17_df = state17_df[['ST', 'STATE', 'OCC_CODE', 'OCC_TITLE', 'OCC_GROUP', 
                           'TOT_EMP', 'JOBS_1000', 'A_MEAN', 'MEAN_PRSE','A_MEDIAN']]
state17_df.head()
```

```python
#delete the rows that the employment estimate is not available (**), replace with na and drop na
state17_df = state17_df.replace("**", np.nan)
state17_df = state17_df.dropna(how="any")

state17_df.count()
```

```python
state17_df.dtypes
```

```python
#creat variables to hold the parameters for filter occupations needed
map_occ_code = "15-0000"
map_occ_tit = "Computer and Mathematical Occupations"

#filter the defined occupation catergory
occ_map_df = state17_df.loc[state17_df['OCC_CODE'] == map_occ_code, :]
occ_map_df.head()
```

```python
occ_map_df.to_csv("images/occ_map_df.csv", index=False, encoding="utf-8", header=True)
```

```python
scl = [
    [0.0, 'rgb(242,240,247)'],
    [0.2, 'rgb(218,218,235)'],
    [0.4, 'rgb(188,189,220)'],
    [0.6, 'rgb(158,154,200)'],
    [0.8, 'rgb(117,107,177)'],
    [1.0, 'rgb(84,39,143)']
]
```

```python
#create text to show infor for each state on the map
occ_map_df['text'] = occ_map_df['STATE'] + '<br>' + \
'Annual Median Wage: '+ '<br>' + \
'$' + occ_map_df['A_MEDIAN'].astype(str)
occ_map_df.head()
```

```python
#use plotly to generate heapmap for the number of employment
    
data = [go.Choropleth(
    colorscale = scl,
    autocolorscale = False,
    locations = occ_map_df['ST'],
    z = occ_map_df['TOT_EMP'].astype(float),
    locationmode = 'USA-states',
    text = occ_map_df['text'],
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(255,255,255)',
            width = 2
        )),
    colorbar = go.choropleth.ColorBar(
        title = "Number of Employment")
)]

layout = go.Layout(
    title = go.layout.Title(
        text = f'2017 US Employment of {map_occ_tit}'
    ),
    geo = go.layout.Geo(
        scope = 'usa',
        projection = go.layout.geo.Projection(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'rgb(255, 255, 255)'),
)

fig = go.Figure(data = data, layout = layout)
plotly.offline.iplot(fig, filename = 'Occupation_map')
```

```python
#use plotly to generate heapmap for the annual median wage of the occupation
data_wage = [go.Choropleth(
    colorscale = scl,
    autocolorscale = False,
    locations = occ_map_df['ST'],
    z = occ_map_df['A_MEDIAN'].astype(float),
    locationmode = 'USA-states',
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(255,255,255)',
            width = 2
        )),
    colorbar = go.choropleth.ColorBar(
        title = "Annual Median Wage ($)")
)]

layout_wage = go.Layout(
    title = go.layout.Title(
        text = f'2017 US Employment of {map_occ_tit}'
    ),
    geo = go.layout.Geo(
        scope = 'usa',
        projection = go.layout.geo.Projection(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'rgb(255, 255, 255)'),
)

fig_wage = go.Figure(data = data_wage, layout = layout_wage)
plotly.offline.iplot(fig_wage, filename = 'Occupation_map_wage')
```

```python
#pie chart for the number of employment by major occupation

#group by occ_code and calculate the nation data (use the agg function)
aggreations = {'TOT_EMP':"sum"}
major_df = state17_df[state17_df['OCC_GROUP'] == "major"].groupby(
           ['OCC_CODE','OCC_TITLE'],as_index=False).agg(aggreations) #(index=False to make CODE and TITLE columnes not index column, for pie label purpose)            
major_df
```

```python
#group by occ_code and calculate the nation data (use the agg function)
aggreations = {'TOT_EMP':"sum"}
major_df = state17_df[state17_df['OCC_GROUP'] == "major"].groupby(['OCC_TITLE']).agg(aggreations)

#sort the data based on employment number
major_sort = major_df.sort_values('TOT_EMP', ascending=False)

#make horizon bar
major_bar = major_sort.plot(kind="barh",figsize=(8, 10), 
                            color="#86bf91", zorder=2, width=0.85, legend=False)
plt.gca().invert_yaxis()

# Despine (box invisible)
major_bar.spines['right'].set_visible(False)
major_bar.spines['top'].set_visible(False)
major_bar.spines['bottom'].set_visible(False)


# Switch off ticks
major_bar.tick_params(axis="both", which="both", bottom="off", top="off", 
                      labelbottom="on", left="off", right="off", labelleft="on")

# Draw vertical axis lines
vals = major_bar.get_xticks()
for tick in vals:
      major_bar.axvline(x=tick, linestyle='dashed', alpha=0.5, color='#eeeeee', zorder=1)
        
#set individual bar lables
rank = 0
for i, v in enumerate(major_sort['TOT_EMP']):
    rank = rank + 1
    major_bar.text(v + 3, i + 0.15, str(f"#{rank}: {v}"), color='white', fontweight='bold')
        
        
# Set x-axis label
major_bar.set_xlabel("Employment", labelpad=20, size=12, fontweight='bold')    
# Set y-axis label
major_bar.set_ylabel("Major Occupations", labelpad=20, size=12, fontweight='bold')
# Set title
major_bar.set_title("2017 US Total Employed Persons by Occupations", 
                    size=12, loc="left", fontweight='bold')
#Set background
plt.style.use('dark_background')


#save the figure
plt.savefig("images/major_bar_2017.png",bbox_inches = 'tight')
plt.show()
```

```python
#group by occ_code and calculate the nation data (use the agg function)
aggreations = {'TOT_EMP':"sum"}
computer_df = state17_df[(state17_df['OCC_CODE'].str.contains("15-1")) | 
                         (state17_df['OCC_CODE'].str.contains("15-2"))].groupby(['OCC_TITLE']).agg(aggreations) 

#sort the data based on employment number
computer_sort = computer_df.sort_values('TOT_EMP', ascending=True)

#make horizon bar
computer_bar = computer_sort.plot(kind="barh",figsize=(8, 10), 
                                  color="#86bf91", zorder=2, width=0.85,legend=False)

# Despine
computer_bar.spines['right'].set_visible(False)
computer_bar.spines['top'].set_visible(False)
computer_bar.spines['bottom'].set_visible(False)

# Switch off ticks
computer_bar.tick_params(axis="both", which="both", bottom="off", top="off", 
                         labelbottom="on", left="off", right="off", labelleft="on")

# Draw vertical axis lines
vals = computer_bar.get_xticks()
for tick in vals:
    computer_bar.axvline(x=tick, linestyle='dashed', alpha=0.5, color='#eeeeee', zorder=1)

# Set x-axis label
computer_bar.set_xlabel("Employment", labelpad=20, size=12, fontweight='bold')
# Set y-axis label
computer_bar.set_ylabel("Detailed Occupations", labelpad=20, size=12, fontweight='bold')
# Set title
computer_bar.set_title("2017 US Employed Persons by Computer and Mathematical Occupations", 
                       size=12, loc="left", fontweight='bold')

plt.style.use('dark_background')

#save the figure
plt.savefig("images/computer_bar_2017.png",bbox_inches = 'tight')
plt.show()
```
